# CREATION DB


```python
# import
import random
import pandas as pd
import numpy as np
from passlib.hash import argon2

from sqlalchemy import Column, Integer, String, Boolean, DateTime, ForeignKey, create_engine, Table
from sqlalchemy.orm import relationship, declarative_base, sessionmaker
from sqlalchemy.sql import func

```

```python
# Creation structure de base
Base = declarative_base()
```

```python
#Class Sitation
class Quote(Base):
    __tablename__ = "quotes"
    quote_id = Column(Integer, primary_key=True)
    quote_text = Column(String)
```

```python
# Initialisation
engine = create_engine("sqlite:///../data/DB_Citation.db")
SesseonLocal = sessionmaker(bind=engine)
Base.metadata.create_all(engine)
```

```python
# Utilisation
Session = sessionmaker(bind=engine)
session = Session()
```

```python
# Ramplir Citation
Nb_quote = 15

# Generation des citations
SUBJECTS = ["La vie", "Le courage", "L'esprit", "La chance", "Le travail", "L'avenir"]
VERBS = ["est", "devient", "reste", "semble", "demeure"]
COMPLEMENTS = [
    "un mystère à explorer",
    "une force inépuisable",
    "une porte vers le changement",
    "une aventure sans fin",
    "un défi qui nous construit"
]

def generate_quote():
    return f"{random.choice(SUBJECTS)} {random.choice(VERBS)} {random.choice(COMPLEMENTS)}."

for i in range(Nb_quote):
    obj = Quote(quote_text=generate_quote())
    session.add(obj)

session.commit() 
```

```python
# Lire le DB
query = session.query(Quote)
df = pd.read_sql(query.statement, session.bind, index_col="quote_id")
df
```

```python
# Fermer le session
session.close()
engine.dispose()
```
